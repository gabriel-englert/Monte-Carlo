```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
plt.style.use('ggplot')
```

```python
t = 1000000
```

Importing data from binomial distribution

```python
n, p = np.loadtxt('bin100.dat', unpack=True)
n, p = np.insert(n,0,0.0), np.insert(p,0,0.0)
```

```python
def f(N,t):
    A = list(range(N))
    t_list = list(range(t))
    NA_list = []
    n_list = [0 for i in range(N)]
    for i in range(t):
        r = rand.randrange(0,N-1,1)
        if r in A:
            A.remove(r)
        else:
            A.append(r)
        #apenas o regime estacionÃ¡rio importa:
        if i>1000:
            NA = len(A)
            NA_list.append(NA)
    for i in NA_list:
      n_list[i] += 1

    n_list = np.asarray(n_list)

    return n_list/t
```

```python
plt.scatter(list(range(100)),f(100,t), label = 'MÃ©todo numÃ©rico')
plt.plot(n,p)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.xlim(25,75)
plt.ylim(0,0.1)
plt.title('ComparaÃ§Ã£o do mÃ©todo numÃ©rico com a distribuiÃ§Ã£o binomial da probabilidade de haver n bolas na urna A')
plt.rcParams['axes.titlepad'] = 20
plt.legend()
plt.show()
```
