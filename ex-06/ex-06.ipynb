**Exercício 6:** Calcule o valor de π por amostragem usando o
método do círculo discutido acima. Faça os mesmos gráficos do problema anterior.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.style.use('ggplot')
```

```python
def random_generator():
  r1 = np.random.uniform(0,1)
  r2 = np.random.uniform(0,1)
  return [r1,r2]
```

```python
t = 100000
i = 0
n = 0
N = 0
pi = 0
N_list = []
pi_list = []
delta = []
while i<t:
  r = random_generator()
  #condição para estar abaixo da curva
  if r[0]**2 + r[1]**2<1:
    n+=1
  N+=1
  N_list.append(N)
  pi = 4*n/N
  pi_list.append(pi)
  i+=1
  delta.append(np.pi-pi)
plt.plot(N_list,pi_list)
constpi_list = [np.pi for i in range(len(N_list))]
zero_list = [0 for i in range(len(N_list))]
y = [2.8,3.0,np.pi,3.2,3.4]
ytick = [2.8,3.0,r'$\pi$',3.2,3.4]
plt.yticks(y, ytick)
plt.xlim(0,len(N_list))
plt.ylim(3,3.2)
plt.plot(N_list,constpi_list)
plt.ylabel(r'$\pi$')
plt.xlabel('N')
plt.title('Estimativa de $\pi$ em função do número (N) de pontos')
plt.rcParams['axes.titlepad'] = 20
plt.savefig('grafico_pi.png')
plt.show()
plt.plot(N_list,delta)
plt.xlim(0,len(N_list))
plt.ylim(-0.05,0.05)
plt.plot(N_list,zero_list)
plt.ylabel('$\delta$')
plt.xlabel('N')
plt.title('Diferença($\delta$) entre o valor analítico de $\pi$ e a estimativa em função do número (N) de pontos')
plt.rcParams['axes.titlepad'] = 20
plt.savefig('grafico_erro.png')
plt.show()
```
