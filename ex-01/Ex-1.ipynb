```python
import matplotlib.pyplot as plt
import numpy as np
import random as rand
plt.style.use('ggplot')
```

```python
t = 500
t_list = list(range(t))
```

Curva analítica, retorna uma lista do NA(Número de bolas na urna A) para cada passo de 0 a t, a partir da equação analítica

```python
def fNA(N,t):
    NA_list= []
    for i in range(t):
        n = N/2+(N*(1-2/N)**i)/2
        NA_list.append(n)
    return NA_list
```

Amostragem

```python
def f(N,t):
    A = list(range(N))
    t_list = list(range(t))
    NA_list = []
    for i in range(t):
        NA = len(A)
        NA_list.append(NA)
        r = rand.randrange(0,N-1,1)
        if r in A:
            A.remove(r)
        else:
            A.append(r)

    plt.plot(t_list,NA_list, label = 'N = ' + str(N))
    return
```

Plots


```python
f(100,t)
f(200,t)
plt.plot(t_list,fNA(100,t),label='NA = 100')
plt.plot(t_list,fNA(200,t),label='NA = 200')
plt.xlabel('t')
plt.ylabel('N')
plt.xlim(0,500)
y = np.arange(0,200,50)
y1 = np.insert(y,3,50)
plt.yticks(y1)
plt.rcParams['axes.titlepad'] = 20
plt.title('Comparação da amostra com a curva análitica do número de bolas na urna A')
plt.legend()
plt.savefig('Gráfico_ex1.png',bbox_inches='tight')
plt.show()
```
