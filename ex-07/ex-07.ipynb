**Exercício 7:** Estime o valor da integral I gerando pontos
aleatórios r1, r2 no retângulo 0 ≤ x, y ≤ 1 e medindo a fração
deles que está abaixo da curva, r2 < f (r1).
Faça o gráfico da estimativa em função do número de pontos
utilizados e compare com o valor conhecido I = 0.673

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.style.use('ggplot')
```

```python
def random_generator():
  r1 = np.random.uniform(0,1)
  r2 = np.random.uniform(0,1)
  return [r1,r2]
```

```python
def f(x):
  return (np.sin(1/x))**2
```

```python
t = 100000
i = 0
n = 0
N = 0
I = 0
N_list = []
I_list = []

while i<t:
  r = random_generator()
  #condição para estar abaixo da curva
  if r[1] < f(r[0]):
    n+=1
  N+=1
  N_list.append(N)
  I = n/N
  I_list.append(I)
  i+=1

plt.plot(N_list,I_list)
constI_list = [0.673 for i in range(len(N_list))]
plt.plot(N_list,constI_list)
plt.ylim(0.66,0.68)
plt.xlim(0,t)
plt.ylabel('I')
plt.xlabel('N')
y = [0.66,0.673,0.68]
ytick = [0.66,0.673,0.68]
plt.yticks(y, ytick)
plt.title('Estimativa da integral de f(x)')
plt.rcParams['axes.titlepad'] = 20
plt.savefig('grafico_estimativa_f.png',)
plt.show()
```
