**Exercício 5:** O problema das agulhas de Buffon e o valor de
π. Considere o gráfico anterior com a curva de sin θ entre 0 e
π e estime sua área como descrito no texto. Faça o gráfico:  
a)do valor obtido para π (compare com o valor conhecido)  
b)da diferença entre essa estimativa e o valor conhecido, ambos
em função do número N de agulhas lançadas.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.style.use('ggplot')
```

```python
def random_generator():
  r1 = np.random.uniform(0,1)*np.pi
  r2 = np.random.uniform(0,1)
  return [r1,r2]
```

```python
t = 100000
i = 0
n = 0
N = 0
pi = 0
N_list = []
pi_list = []
delta = []
while i<t:
  r = random_generator()
  #condição para estar abaixo da curva
  if r[1]<np.sin(r[0]):
    n+=1
  N+=1
  N_list.append(N)
  if n == 0:
    pi = 2*N
  else:
    pi = 2*N/n
  pi_list.append(pi)
  i+=1
  delta.append(np.pi-pi)
plt.plot(N_list,pi_list)
constpi_list = [np.pi for i in range(len(N_list))]
zero_list = [0 for i in range(len(N_list))]
y = [2.8,3.0,np.pi,3.2,3.4]
ytick = [2.8,3.0,r'$\pi$',3.2,3.4]
plt.yticks(y, ytick)
plt.xlim(0,len(N_list))
plt.ylim(3,3.2)
plt.plot(N_list,constpi_list)
plt.ylabel(r'$\pi$')
plt.xlabel('N')
plt.title('Estimativa de $\pi$ em função do número (N) de agulhas')
plt.rcParams['axes.titlepad'] = 20
plt.savefig('grafico_pi',bbox_inches='tight')
plt.show()
plt.plot(N_list,delta)
plt.xlim(0,len(N_list))
plt.ylim(-0.05,0.05)
plt.plot(N_list,zero_list)
plt.ylabel('$\delta$')
plt.xlabel('N')
plt.title('Diferença($\delta$) entre o valor analítico de $\pi$ e a estimativa em função do número (N) de agulhas')
plt.rcParams['axes.titlepad'] = 20
plt.savefig('grafico_erro',bbox_inches='tight')
plt.show()
```
